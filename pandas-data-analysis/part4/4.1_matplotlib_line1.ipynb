```python
# 라이브러리 불러오기
import pandas as pd
import matplotlib.pyplot as plt

# Excel 데이터를 데이터프레임 변환 
df = pd.read_excel('./data/시도별_전출입_인구수.xlsx')

# 첫 5행 출력
df.head()
```

```python
#nan Nan 결측치 -- 데이터 없음
# 1. 제거 - 데이터가 충분홰서 해당 결측치를 해당 데이터를 삭제해도 충분할때...
# 2. 대처
    # 단순실수로 누락되어서 뻔히 어떤값인지 알 수 있을때
    # 결측치가 전체데이터 대비 적다.. 대략 15% 미만이고, 데이터가 적어서 살려야할때 -  평균, 표준편차, 빈도수등 다양하게 채움
    #보간기법이라고 함--> 자체 머신러닝을 이용해서 해당 부분의 값을 예측
```

```python
df['전출지별'].unique()
```

```python
# 누락값(NaN)을 앞 데이터로 채움 (엑셀 양식 병합 부분)
df = df.ffill()
df.head()
```

```python
# 서울에서 다른 지역으로 이동한 데이터만 추출하여 정리
mask = (df['전출지별'] == '서울특별시') & (df['전입지별'] != '서울특별시') 
df_seoul = df[mask]
df_seoul = df_seoul.drop(['전출지별'], axis=1)
df_seoul = df_seoul.rename({'전입지별':'전입지'}, axis=1)
df_seoul = df_seoul.set_index('전입지')
df_seoul = df_seoul.drop(index = ['전국'])
df_seoul.T.drop(columns= ['경기도']).plot()
```

```python
# 한글 폰트설정
import matplotlib.pyplot as plt
import matplotlib as mpl

mpl.rcParams['font.family'] = 'Malgun Gothic'
mpl.rcParams['axes.unicode_minus'] = False
```

```python
# 다른지역에서 서울로 이동한 데이터
mask = (df['전출지별'] != '서울특별시') & (df['전입지별'] == '서울특별시') 
df_seoul = df[mask]
df_seoul = df_seoul.drop(columns=['전입지별'])
df_seoul = df_seoul.rename(columns={'전출지별':'전출지'})
df_seoul = df_seoul.set_index('전출지')
df_seoul = df_seoul.drop(index=['전국'])
df_seoul.T.drop(columns= ['경기도']).plot()
#시각화
df_seoul.T.plot()
# df_seoul.T.plot()
```

```python
# 서울에서 경기도로 이동한 인구 데이터 값만 선택 
sr_one = df_seoul.loc['경기도']
sr_one.head()
```

```python
# x, y축 데이터를 plot 함수에 입력
plt.plot(sr_one.index, sr_one.values)
```

```python
# 판다스 객체를 plot 함수에 입력
plt.plot(sr_one)
plt.show()
```

```python
import gdown

file_id = '1oxUlhFbbJqUEdAZnynTJUav5mCkM9sb7'
url = f'http://drive.google.com/uc?id={file_id}'
output = 'credit.csv'

gdown.download(url,output,quiet=False)
df = pd.read_csv(output)
df.head()

#installments 할부개월수
#days_of_week 월요일이 0 일요일이 6
#holyday 1이면 공휴일 0이면 공휴일이 아님
```

```python
#데이터가 날짜인데.. 문자열 형테 즉 object이면
#datetime으로 형식을 변경해서 년월일  을 분리할 수 있다.
df['datetime'] = pd.to_datetime(df['date'])
df['year'] = df['datetime'].dt.year
df['month'] = df['datetime'].dt.month
df.head()
```

```python
# 년도별 신용카드 매출액 구해서 출력
temp = df.groupby('year')[['amount']].sum()
print(temp)
ax = temp.plot(kind='bar')
# 시각화 plot x축의 시간정보는 오름차순으로 정렬되었는지 확인


```
