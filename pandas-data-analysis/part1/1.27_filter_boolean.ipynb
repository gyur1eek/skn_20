```python
# 라이브러리 불러오기
import pandas as pd
import seaborn as sns

# # titanic 데이터셋 로딩
titanic = sns.load_dataset('titanic').rename(columns={'sex':'gender'})

# # 행 인덱스 0~9 범위에서 age, fare 2개 열을 선택하여 데이터프레임 만들기
# titanic = titanic.loc[0:9, ['age','fare']]

# # 데이터프레임 출력
titanic.head(3)
titanic_simple = titanic.drop(columns=['alive','class','who','embark_town','deck'])  #inplace : 실제값 바로 바꿔줌
titanic_simple.head(3)
```

```python
# 조건식 적용 (20세 미만: age < 20)
titanic_under20 = titanic[titanic['age'] < 20]
# 생존한 승객
titanic_under20_survived = titanic_under20[titanic_under20['survived'] == 1]
# 여성의 데이터만 출력
titanic[(titanic['survived'] == 1) & (titanic['gender'] == 'female')]
```

```python
# 불린 인덱싱 (20세 미만의 조건을 충족하는 행을 필터링)
# titanic_under20 = titanic[ titanic['age'] < 20]
# titanic_under20[titanic_under20['gender'] == 'female']].head(3)
# age < 20 그리고 gender == 'female'
mask = (titanic['age']<20) & (titanic['gender'] == 'female')  # and로 하면 오류가 남..
titanic[mask]
```

```python
# 불린 인덱싱 (loc 인덱서 활용)
titanic.loc[titanic['age'] < 20]
```

```python
# 불린 인덱싱 (NOT 논리연산)
titanic.loc[~(titanic['age'] < 20)]
```

```python
# 나이가 10대(10~19세)인 승객 (AND 논리연산)
mask1 = (titanic.age >= 10) & (titanic.age < 20)
titanic_teenage = titanic[mask1]
titanic_teenage.head()
```

```python
# 나이가 10세 미만(0~9세)이고 여성인 승객 (AND 논리연산, loc 인덱서)
mask2 = (titanic.age < 10) & (titanic.gender == 'female')
titanic_female_under10 = titanic.loc[mask2]
titanic_female_under10.head()
```

```python
# 나이가 10세 미만(0~9세) 또는 60세 이상인 승객 (OR 논리연산)
# age, gender, alone 열만 선택 (loc 인덱서, fancy indexing)
mask3 = (titanic.age < 10) | (titanic.age >= 60)
titanic_under10_morethan60 = titanic.loc[mask3, ['age', 'gender', 'alone']]
titanic_under10_morethan60.head()
```

```python
# 불린 시리즈를 새로운 열로 추가
titanic['age_under_20'] = titanic['age'] < 20
titanic.head(3)
```

```python
# 1등석과 2등석, 3등석 : pclass
# 선실별 승객수 len(데이터프레임)
print( len(titanic[titanic['pclass'] ==1]) )
print( len(titanic[titanic['pclass'] ==2]) )
print( len(titanic[titanic['pclass'] ==3]) )
```
