```python
import pandas as pd
import seaborn as sns
```

```python
mpg = sns.load_dataset('mpg')
mpg.head(3)
```

```python
# groupby -실린더별  마력의 평균
    # name 별 연식의 평균
    # name이 공백을 기준으로 제조사[0], 모델
    # 제조사 컬럼을 만들고 제조사를 추가

# 필터-mpg가 15이상인 모델
# index를 재구성 - origin으로 구성
#iloc, loc을 이용해서 데이터를 row, col 다양하게 출력
```

```python
# groupby -실린더별  마력의 평균
mpg.groupby("cylinders")['horsepower'].mean()
# name별 연식의 평균
mpg.groupby("name")['model_year'].mean()

# mpg['maker'] = [i[0] for i in mpg['name'].str.split().values]
mpg['make'] = mpg['name'].apply(lambda x: x.split()[0])
mpg.head()

```

```python
#필터 - mpg가 15이상인 모델
mpg = mpg.query("mpg>=15")
```

```python
mpg.groupby('make')['name'].count()
```
